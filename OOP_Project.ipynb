```python
class NLP:

    # Private class variable to store user information
    __database = {}

    def __init__(self):
        self.main_menu()

    def main_menu(self):
        """
        Display the main menu and handle user choices for login, registration, or exit.
        """
        choice = input("""
        ****** Menu ******
        1. Login
        2. Not a member? Register
        3. Exit
        Choose an option: """)

        if choice == '1':
            self.login()

        elif choice == '2':
            self.register()

        elif choice == '3':
            exit()

    def register(self):
        """
        Handle user registration by collecting name, email, and password, and adding them to the database.
        """
        name = input("Enter Your Name: ")
        email = input("Enter Your Email: ")
        password = input("Create a password for your NLP-Services account: ")

        if email in NLP.__database:
            print("Email already exists. Please log in.")
            self.main_menu()

        else:
            NLP.__database[email] = [name, password]
            print("Registered Successfully")
            self.main_menu()

    def login(self):
        """
        Handle user login by verifying email and password, granting access if valid.
        """
        email = input("Enter email: ")
        password = input("Enter password: ")

        if email in NLP.__database:
            if password == NLP.__database[email][1]:
                print("****** Welcome ******")
                self.second_menu()
            else:
                print('Incorrect password')
                self.main_menu()
        else:
            print('Email not registered. Please register first.')
            self.main_menu()

    def second_menu(self):
        """
        Display the secondary menu and handle user choices for NER, Sentiment Analysis, Emotion Prediction, or Logout.
        """
        choice = input("""
        ****** Secondary Menu ******
        1. Named Entity Recognition (NER)
        2. Sentiment Analysis
        3. Emotion Prediction
        4. Logout
        Choose an option: """)

        if choice == '1':
            self.ner()

        elif choice == '2':
            self.sentiment_analysis()

        elif choice == '3':
            self.emotion_prediction()

        elif choice == '4':
            exit()

    def ner(self):
        """
        Perform Named Entity Recognition on user-provided text and display the results.
        """
        import paralleldots
        paralleldots.set_api_key('Replace this code with Your Komprehend Key')
        text = input("Enter some text for Named Entity Recognition: ")
        response = paralleldots.ner(text)
        print(response)
        self.second_menu()

    def sentiment_analysis(self):
        """
        Perform Sentiment Analysis on user-provided text and display the results.
        """
        import paralleldots
        paralleldots.set_api_key('Replace this code with Your Komprehend Key')
        text = input("Enter some text for Sentiment Analysis: ")
        response = paralleldots.sentiment(text)
        print(response)
        self.second_menu()

    def emotion_prediction(self):
        """
        Perform Emotion Prediction on user-provided text and display the results.
        """
        import paralleldots
        paralleldots.set_api_key('Replace this code with Your Komprehend Key')
        text = input("Enter text for Emotion Prediction: ")
        result = paralleldots.emotion(text)
        print(result)
        self.second_menu()
```
